````python
# CSV
file_path = "/content/drive/MyDrive/aiffel_final_project/data_renew/book_data.csv"
df = pd.read_csv(file_path)
print(df.columns)
df = df.sample(n=10000, random_state=2025)
````

````python
# 텍스트 청크 분할
def split_text(text, chunk_size=1000, overlap=100):
    chunks = []
    for i in range(0, len(text), chunk_size - overlap):
        chunks.append(text[i:i + chunk_size])
    return chunks

# RAG_DB 생성: 각 책의 텍스트(행별)를 청킹하여 리스트 저장
RAG_DB = []
for index, row in df.iterrows():
    title = row['TITLE_NM']
    authr = row['AUTHR_NM']
    price = row['PRC_VALUE']
    text = row['BOOK_INTRCN_CN']
    chunks = split_text(text)
    for chunk in chunks:
        RAG_DB.append({
            'Title': title,
            'Authr': authr,
            'Price': price,
            'Text': chunk
        })
# -> 1행과 2행의 서로다른 책은 각 길이에 상관없이 다른 청크로 들어가도록

# RAG_DB - LangChain의 Document
# LangChain Document 생성 (metadata는 추후, 필요시 확장 가능)
documents = [
    Document(
        page_content=f"Title: {entry['Title']}\nAuthr: {entry['Authr']}\nPrice: {entry['Price']}\nText: {entry['Text']}"
    ) for entry in RAG_DB
]
````

````python
# -------------------------------------------------------------------
# [MODULE 2] Embedding & Vectorstore (변경 없음, 모듈 내부에 포함)
# -------------------------------------------------------------------
# HuggingFace 임베딩 모델
model_name = "intfloat/multilingual-e5-large-instruct"
hf_embeddings = HuggingFaceEmbeddings(
    model_name=model_name,
    model_kwargs={"device": "cuda"},  # GPU 사용
    encode_kwargs={"normalize_embeddings": True},
)
````

````python
# 문서 임베딩 및 FAISS 벡터스토어 생성
text_embedding_pairs = []
metadata_list = []
for doc in tqdm(documents, desc="Processing Documents", unit="document"):
    embedding = hf_embeddings.embed_query(doc.page_content)
    text_embedding_pairs.append((doc.page_content, embedding))
    metadata_list.append(doc.metadata)

vectorstore = FAISS.from_embeddings(
    text_embeddings=text_embedding_pairs,
    metadatas=metadata_list,
    embedding=hf_embeddings
)
````

````python
# FAISS retriever 생성 (검색할 문서 개수 Top k=5)
dense_retriever = vectorstore.as_retriever(search_kwargs={"k": 5})
````

````python
# -------------------------------------------------------------------
# [MODULE 3] LLM & Prompt 설정 (프롬프트 로직은 그대로 유지)
# -------------------------------------------------------------------
# Gemini-1.5-flash 모델
llm_gemini = ChatGoogleGenerativeAI(model="gemini-1.5-flash", temperature=0.0)

# RetrievalQA 체인 (기존 dpr 기반 retrieval)
dpr_qa_chain = RetrievalQA.from_chain_type(
    llm=llm_gemini,
    retriever=dense_retriever,
    return_source_documents=True
)
````

````python
# robust_parse_json_response Json 응답 파싱용
# -------------------------------------------------------------------
# [MODULE 4] Utility 함수들 (파싱 등)
# -------------------------------------------------------------------
def robust_parse_json_response(response):
    """
    Gemini 응답을 JSON으로 파싱 (dict면 그대로, 문자열이면, 코드블록 제거 후 파싱)
    """
    if isinstance(response, dict):
        if "text" in response and isinstance(response["text"], str):
            response_text = response["text"]
        else:
            return response
    else:
        response_text = response
    response_text = response_text.strip()
    if response_text.startswith("```json"):
        response_text = response_text[len("```json"):].strip()
    if response_text.endswith("```"):
        response_text = response_text[:-3].strip()
    try:
        data = json.loads(response_text)
        return data
    except Exception as e:
        print("JSON 파싱 에러:", e)
        return None
````

````python
# robust_parse_llm_response 함수 (멀티턴 프롬프트 free text 파싱용)
def robust_parse_llm_response(response_text):
    """
    LLM 응답 텍스트에서 검색 확률, 검색 쿼리, 추가 질문 추출
    예시 형식:
    1. 검색 확률: 0.8
    2. 검색 쿼리: "AI 철학 관련 최신 도서"
    3. 추가 질문: "AI 철학 관련해서 어떤 주제가 궁금하세요?"
    """
    cleaned_text = re.sub(r'\*\*', '', response_text)
    search_score = None
    search_query = None
    follow_up_question = ""
    score_match = re.search(r"검색\s*확률\s*[:：]\s*([\d\.]+)", cleaned_text)
    if score_match:
        try:
            search_score = float(score_match.group(1))
        except Exception as e:
            print("검색 확률 파싱 에러:", e)
    query_match = re.search(r"검색\s*쿼리\s*[:：]\s*\"(.*?)\"", cleaned_text)
    if query_match:
        search_query = query_match.group(1).strip()
    follow_match = re.search(r"추가\s*질문\s*[:：]\s*\"(.*?)\"", cleaned_text)
    if follow_match:
        follow_up_question = follow_match.group(1).strip()
    return search_score, search_query, follow_up_question
````

````python
# 사용자 선호도 수집
user_preferences = defaultdict(list)
log_history = []  # 로그 저장 옵션
def categorize_preference(question, response):
    if "장르" in question or "어떤 책" in question:
        user_preferences["genre"].append(response)
    elif "작가" in question or "좋아하는 작가" in question:
        user_preferences["author"].append(response)
    elif "목적" in question or "이유" in question:
        user_preferences["purpose"].append(response)
    else:
        user_preferences["misc"].append(response)
````

````python
# -------------------------------------------------------------------
# [MODULE 5] PreRetrieval Module (멀티턴 프롬프트 관련 로직을 캡슐화(모듈화))
# -------------------------------------------------------------------
class PreRetrievalModule:
    def __init__(self, search_query_chain, final_query_chain, threshold=0.6): # threshold!!
        self.search_query_chain = search_query_chain
        self.final_query_chain = final_query_chain
        self.threshold = threshold

    def get_search_decision(self, query, query_history):
        # 최근 5개 대화 요약 (query_history 활용)
        query_summary = "\n".join(query_history[-5:])
        response = self.search_query_chain.invoke({
            "history": query_summary,
            "query": query,
            "if": "✅ 검색이 가능한 경우:",
            "else": "❌ 아직 검색이 불가능한 경우:",
            "impormation": "📌 사용자 선호도 분석:",
            "strategy": "🔍 검색 전략:",
        })
        response_text = response["text"].strip()
        print("\n[🔍 멀티턴 프롬프트 응답]\n", response_text)
        search_score, base_search_query, follow_up_question = robust_parse_llm_response(response_text)
        print(f"\n[디버그] 파싱 결과: 검색 확률={search_score}, 기본 검색 쿼리='{base_search_query}', 추가 질문='{follow_up_question}'")
        return search_score, base_search_query, follow_up_question

    def generate_final_query(self, query_history, fallback):
        # Final Query Chain를 활용하여 최종 검색 쿼리 생성 - 추후 변경하려면, Final Query chain 부분을 변경 후, 부착시키기
        final_query_result = self.final_query_chain.invoke({
            "history": "\n".join(query_history),
            "fallback": fallback
        })
        if isinstance(final_query_result, dict):
            final_search_query = final_query_result.get("text", "").strip()
        else:
            final_search_query = final_query_result.strip()
        print(f"\n[🔎 최종 검색 쿼리 생성]: {final_search_query}")
        return final_search_query
````

````python
# -------------------------------------------------------------------
# [MODULE 6] Retrieval Module (기존 DPR‑QA 체인 모듈화)
# -------------------------------------------------------------------
class RetrievalModule:
    def __init__(self, dpr_qa_chain):
        self.dpr_qa_chain = dpr_qa_chain

    def retrieve(self, search_query):
        result = self.dpr_qa_chain.invoke(search_query)
        initial_answer = result['result']
        retrieved_docs = result['source_documents']
        return initial_answer, retrieved_docs
````

````python
# -------------------------------------------------------------------
# [MODULE 7] PostRetrieval Module (Gemini 기반 post processing 모듈화)
# -------------------------------------------------------------------
class PostRetrievalModule:
    def __init__(self, llm, post_retrieval_prompt):
        self.llm = llm
        self.post_retrieval_prompt = post_retrieval_prompt

    def post_process(self, query, retrieved_docs):
        docs_text = "\n\n".join([doc.page_content for doc in retrieved_docs])
        post_retrieval_chain = LLMChain(llm=self.llm, prompt=self.post_retrieval_prompt)
        post_ret_response = post_retrieval_chain.invoke({
            "query": query,
            "documents": docs_text
        })
        print("\n[🔄 Gemini Post‑Retrieval 응답]\n", post_ret_response)
        post_data = robust_parse_json_response(post_ret_response)
        if post_data is not None:
            final_answer = post_data.get("final_answer", "").strip()
            sources = post_data.get("sources", [])
        else:
            final_answer = ""
            sources = []
        return final_answer, sources
````

````python
# -------------------------------------------------------------------
# [MODULE 8] Modular RAG 메인 클래스 (모듈 통합 및 전체 프로세스 실행)
# -------------------------------------------------------------------
class ModularRAG:
    def __init__(self, pre_module, retrieval_module, post_module, threshold=0.6):
        self.pre_module = pre_module
        self.retrieval_module = retrieval_module
        self.post_module = post_module
        self.threshold = threshold

    def process_query(self, query, query_history):
        # 기존 search_and_generate_answer 함수의 로직을 모듈로 통합
        while True:
            search_score, base_search_query, follow_up_question = self.pre_module.get_search_decision(query, query_history)
            if search_score is None:
                print("\n[❌ 응답 파싱 실패: 추가 정보 필요]")
                extra_info = input("추가 정보를 입력해주세요: ")
                query_history.append(f"사용자(추가): {extra_info}")
                query = f"{query} {extra_info}"
                continue

            # threshold 조건 충족 시 Advanced RAG 기법(pre&post retrival) 실행
            if search_score >= self.threshold and base_search_query:
                final_search_query = self.pre_module.generate_final_query(query_history, base_search_query)
                initial_answer, retrieved_docs = self.retrieval_module.retrieve(final_search_query)
                final_answer, sources = self.post_module.post_process(query, retrieved_docs)
                if not final_answer:
                    final_answer = initial_answer.strip()
                    sources = [doc.metadata.get('Title', '출처 없음') for doc in retrieved_docs]
                if sources:
                    book_info = "\n".join([f"- {title}" for title in sources])
                    answer_with_info = f"{final_answer}\n\n[📚 책 정보]\n{book_info}"
                    print("\n[📚 책 정보]\n", book_info)
                else:
                    answer_with_info = final_answer
                return answer_with_info

            # 보충 질문이 있을 경우 (추가 정보 요청) - 변경사항 X
            if follow_up_question:
                print(f"\n[🤖 보충 질문: {follow_up_question}]")
                query_history.append(f"AI: {follow_up_question}")
                user_response = input("\n사용자 응답: ")
                query_history.append(f"사용자: {user_response}")
                categorize_preference(follow_up_question, user_response)
                print("\n[📚 사용자 선호도 업데이트 완료!]")
                query = f"{query} {follow_up_question} {user_response}"
                continue

            # 검색 확률이 thresold 보다 아래거나, 기본 검색 쿼리 부재 시 추가 정보 모듈로 넘어가서 추가 정보 생성
            if search_score < self.threshold or not base_search_query:
                print("\n[❌ 검색 확률 낮거나 검색 쿼리 없음: 추가 정보 필요]")
                extra_info = input("추가 정보를 입력해주세요: ")
                query_history.append(f"사용자(추가): {extra_info}")
                query = f"{query} {extra_info}"
                continue
````

````python
# -------------------------------------------------------------------
# [MODULE 9] Interactive 멀티턴 대화 실행 함수 모듈
# -------------------------------------------------------------------
def interactive_multi_turn_qa():
    query_history = []  # 실행마다 초기화 (사용자 초기화)
    # 모듈 인스턴스 생성 (모듈화된 구조로 통합)
    pre_module = PreRetrievalModule(search_query_chain, final_query_chain)
    retrieval_module = RetrievalModule(dpr_qa_chain)
    post_module = PostRetrievalModule(llm_gemini, post_retrieval_prompt)
    modular_rag = ModularRAG(pre_module, retrieval_module, post_module)

    while True:
        clear_output(wait=True)
        print("📚 멀티턴 AI 기반 책 추천 시스템 (종료하려면 'quit' 입력)")
        print("-" * 50)
        query = input("질문을 입력하세요: ")
        if query.lower() == 'quit':
            print("\n[📝 대화 저장 중...]")
            log_history.append(query_history)
            print("대화가 저장되었습니다. 프로그램을 종료합니다.")
            break
        query_history.append(f"사용자: {query}")
        answer = modular_rag.process_query(query, query_history)
        print("\n[💬 AI의 답변]")
        print(answer)
        query_history.append(f"AI: {answer}")
        input("\n-> 계속하려면 Enter를 누르세요...")
````
